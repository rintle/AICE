### insert 함수를 사용해 "원하는 위치에 칼럼 생성"하기

-loc : 삽입될 열의 위치

-value : 삽입될 열의 값

### 그룹화하기
groupby  
데이터프레임의 같은 값을 하나로 묶어 통계나 집계 결과를 확인하기 위해 사용한다.  
아래 과정을 거쳐 그룹화한다.
- Split : 그룹별로 데이터를 나눈다.
- Apply : 각 그룹별로 집계 함수를 적용한다.
- Combine : 그룹별 집계 결과를 하나로 합친다.

```python
airline_group = flight.groupby('airline')
airline_group
```

```python
airline_group.groups
```

groups로 확인하면 가독성이 구리니, 내부 메소드를 활용해 확인하자.  
- count : 데이터 개수
- size : 집단별 크기
- sum : 데이터의 합
- mean, std, var : 평균, 표준편차, 분산
- min, max : 최솟값, 최댓값

```python
airline_group.count()
```

```python
airline_group.min()
```

```python
# airline_group.mean() 만 하게 되면 오류가 발생한다.
airline_group.mean(numeric_only=True)
```

```python
airline_group.mean(numeric_only=True)[['price']]
```

복수개의 칼럼을 기준으로 groupby를 사용하면 다중 인덱싱 가능

```python
flight.groupby(['airline', 'arrival_time']).mean(numeric_only=True)
# flight.groupby(['airline', 'arrival_time']).min()
```

```python
mul_airline_group = flight.groupby(['airline', 'arrival_time'])
mul_airline_group.mean(numeric_only=True).head(18)
```

AirAsia의 Evening 정보만 확인해보기

웬만해서는 [] 쓸 때 두번 감싸주는게 표기상으로 이쁘게 나오네.

```python
flight.groupby(['airline', 'arrival_time']).mean(numeric_only=True).loc[[('AirAsia', 'Evening')]]
# flight.groupby(['airline', 'arrival_time']).min().loc[[('AirAsia', 'Evening')]]
```

### 인덱스로 그룹화하기

set_index : 칼럼을 인덱스로 변경하는 경우에 사용.   
 기존 인덱스를 제거하고, 칼럼중 하나를 인덱스로 설정  
reset_index : 인덱스를 초기화

```python
flight
```

```python
flight.set_index(['airline', 'arrival_time'])
# 보면, 그냥 순서대로 인덱싱을 한 것에 불과하다.
```

```python
flight.set_index(['airline', 'arrival_time']).groupby(level=[0]).mean(numeric_only=True)
```

```python
flight.set_index(['airline', 'arrival_time']).groupby(level=[0,1]).aggregate([np.min, np.max])
```

## 피벗 테이블 생성하기

## merge/join 활용해 병합하기
merge와 join함수를 이용해 key를 기준으로 데이터 프레임을 병합

```python
customer = pd.DataFrame({'customer_id' : np.arange(6),
                         'name': ['James', 'Elly','Tom','Givert','Aiden','Brody'],
                         '나이' : [40,20,21,30,31,18]})
customer
```

```python
orders = pd.DataFrame({'customer_id': [1,1,2,2,2,3,3,1,4,9],
                       'item': ['마우스','충전기','이어폰','헤드셋','전자펜','키보드','전자펜','마우스','키보드','케이스'],
                       'quantity': [1,2,1,1,3,2,2,3,2,1]})
orders
```

```python
pd.merge(customer,orders, on='customer_id')
```

```python
pd.merge(customer, orders, on='customer_id', how='inner')
```

```python
pd.merge(customer, orders, on='customer_id', how='left')
```

```python
pd.merge(customer, orders, on='customer_id', how='right')
```

```python
pd.merge(customer, orders, on='customer_id', how='outer')
```

```python
customer
```

```python
cust1 = customer.set_index('customer_id')
cust1
```

```python
orders
```

```python
order1 = orders.set_index('customer_id')
order1
```

```python
pd.merge(cust1, order1, left_index=True, right_index=True)
```

```python
pd.merge(cust1, order1, on='customer_id')
```

```python
pd.merge(cust1, order1, on='customer_id', how='right').groupby('item').sum().sort_values(by='quantity', ascending=False)
```

```python
pd.merge(cust1, order1, on='customer_id', how='left')
```

```python
pd.merge(cust1, order1, on='customer_id', how='inner').groupby(['name', 'item']).sum()
```

```python
pd.merge(cust1, order1, on='customer_id', how='inner').groupby(['name', 'item']).sum().loc['Elly']
```

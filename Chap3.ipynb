## 데이터프레임의 기본 정보 확인하기
.shape : 행과 열의 개수를 튜플 형태로 변환해줌

```python
flight.shape
# 300153행, 12열
# column이 12개 있다는 뜻.
```

.columns : Column의 이름을 확인할 때 사용한다

```python
flight.columns
```

.info() : 데이터 유형, 각 칼럼의 데이터 수 등을 한번에 볼 수 있다.  
결측치를 파악하기 용이하다.

```python
flight.info()
```

.describe() : 숫자형 데이터의 통계치를 확인하는데 유용하다.
범주형 데이터의 경우 데이터 개수, 최빈값, 최빈값 개수를 제공한다.

```python
flight.describe()
```

include='all' 파라미터 추가 시, 모든 데이터에 대한 describe를 볼 수 있다.  
위에서 언급한 것 처럼, 숫자형 데이터는 통계치, 범주형 데이터는 데이터 개수와 최빈값을 보여준다.

```python
flight.describe(include='all')
```

.dtypes : 데이터프레임이 갖는 데이터 형태의 종류를 보여준다.

```python
flight.dtypes
```

value_counts() : 선택한 column에 대해서 각 데이터의 발생 횟수를 확인할 수 있다.

```python
flight.source_city.value_counts()
flight['source_city'].value_counts()
flight['price'].value_counts()
```

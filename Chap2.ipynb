## 기존 데이터로 부터 새로운 데이터프레임 구성하기
pd.read_csv로 불러올 때, index를 무엇으로 할지, 어떤 column을 사용할 지 결정지어서 불러올 수 있다.  
usecols에는 index_col을 반드시 포함해야 한다. (포함하지 않는다면 어폐가 생기기 때문.)

```python
flight2 = pd.read_csv('Clean_Dataset.csv', index_col = 'stops', usecols=['stops','departure_time','arrival_time','destination_city'])
flight2
```

## 범주형 데이터 비교 분석하기
crosstab

```python
pd.crosstab(index = flight.source_city, columns = flight.arrival_time)
```
